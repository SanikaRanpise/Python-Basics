Functions

What is function
As the name says, function is something that has a function (i.e. purpose)

In other words, a function must have a clear definition otherwise it cannot function properly!

```python
def hello():
  print('Hello world!')
hello()
```

So what can we say about function now?

A function can be defined as the organized block of reusable code, which can be called whenever required.

Functions start with keyword def followed by function_name() And end with : followed by indentation.

Functions allow reusability of the code as we can call Python functions multiple times anywhere in a program.

To call the function, use the function name followed by the parentheses.



```python
hello()
hello()
```

```python
def hello(name):
  print(f'hello {name}')

hello('Sanika')
hello(21)
```

```python
print(type(hello))
```

```python
def addthis(first,second):
  print(first+second)

addthis(12,13)
```

```python
# If the function definition has arguments, then while calling arguments must be passed, else it'll give error
# Both the calls below will give error because addthis() expects two parameters

# addthis()
# addthis(12)
```

Some more examples

```python
def hello_you(name):
  return(f'hello {name.upper()}')
print(hello_you('students'))
```

Return statement

Almost all the time, function returns some value/data

```python
def hi(name):
  message = 'Hi'  +  name
  return message

name = input("Enter your name: ")
print(hi (name))
```

```python
def add():
  a = 40
  b = 30
  c = a+b
  return c

print("The sum is: ",add())
```

```python
def add(a,b):
    return a+b;

a = int(input("Enter a: "))    # will give error if you pass string here
b = int(input("Enter b: "))

print("Sum = ",add(a,b))
```

Let's write a function to calculate interest on loan

```python
p = int(input("Enter the principal amount: "))
r = float(input("Enter the rate of interest?: ")) # sometimes interest rate could be in fraction like 8.5 or something, so float
t = int(input("Enetr the time in years: "))

def simple_interest(p,t,r):
  return(p*t*r)/100

print("Simple Interest: ",simple_interest(p,t,r))
```

```python
# For readability, you can call function by writing the name of arguments

def find_distance(speed,time):
  print(time,speed)
  return speed * time
d = find_distance(speed = 10 , time = 3)
print(d)
d = find_distance(time=3, speed=10)     # even if order is changed, no problem. This is called KEYWORD ARGUMENTS.
print(d)
d = find_distance(3,10)     # This is not the same as above (in this case, speed=3, time=10)
print(d)
```
