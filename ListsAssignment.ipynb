Iterating a list

```python
countries = ['Russia', 'India', 'US', 'Japan', 'Germany', 'Italy', 'Zambia']
```

```python
for item in countries:
  print(item)
```

Basically, we are telling python that, for every element in countries, print that element

```python
for x in countries:
  print(x)
```

```python
for x in countries:
  print(x)
print('Now IM outside the for loop')

print('----------')

for x in countries:
  print(x)
  print('Now IM inside the loop')
```

```python
for item in countries:
  print(item , len(item))
```

```python
for item in countries:
  print(item , item.upper(), len(item))
```

```python
for item in countries:
  print(f'{item}, { item.upper()},{len(item)}')
```

```python
for item in countries:
  print(f'{item}--- { item.upper()}---{len(item)}')
```

```python
# Use enumerate() function to get index of element.
for id , item in enumerate(countries):
  print(id , item)
```

```python
for index, item in enumerate(countries):    # id can be changed to anyother thing
  print(index , item)
```

```python
# let's try to create a list using for loop and append()
cubes = []     # first define a blank list
for i in range(5):     # you can explore what range(5) gives. In short, it returns numbers upto 5
  cubes.append(i**3)
print(cubes)
```

```python
cubes = []
for i in range(5):
  print("i = ", i)
  cubes.append(i**3)
  print("Number appended i**3 = ",i**3)
  print("Current status of cube list", cubes)
print()
print('Outside for loop')
print(cubes)
```
