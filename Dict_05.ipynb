Dictionary

Basics
Dictionaries are used to store data values in key:value pairs.

A dictionary is a collection which is ordered, changeable and does not allow duplicates.

Dictionaries are written with curly brackets, and have keys and values:

```python
company = {
  "name": "Tesla",              # this is called dictionary iteam. This dictionary has 3 items.
  "founder": 'Elon Musk',
  "established": 2003
}
print(company)
```

Observe the structure. Dictionary started and ended with {}. Within these braces, there is key: value structure. Left side is key, right side is value. (Keys must be of type string, value could be anything - int, str, float, list or another dictionary)

```python
print(type(company))
```

```python
print(company['name'])
```

```python
print(len(company))
```

```python
company['name'] = 'TESLAA'
print(company)
```

```python
company['location'] = 'US'
print(company)
```

Dictionary Functions

```python
help(dict)
```

```python
print(company.get('name'))
print(company.get('namee'))
```

```python
print(company.keys())
print(company.values())
```

```python
company['newkey'] = 'newval'
print(company.keys())
print(company.values())
```

```python
student = {
    'name' : 'John',
    'age' : 12,
    'class' : 7
}

print(student)
student.pop('class')
print(student)
```

```python
# del can also be used to delete an item

student = {
    'name' : 'John',
    'age' : 12,
    'class' : 7
}

print(student)

del student['class']
print(student)
```

```python
# To delete complete dictionary, simply use del
# del student    # Once deleted you can't delete again.
```

```python
# clear() clears the dictionary without deleting the defined object. The dictionary just becomes empty (clear)

student = {
    'name' : 'John',
    'age' : 12,
    'class' : 7
}

print(student)
student.clear()
print(student)
```

Dictionary Looping

```python
# Let's first define dictionary in a different way
new = dict({'name': 'Apple', 'year': 1975, 'founder': 'Steve Jobs and Steve Wozniak'})
print(new)
print(type(new))
```

```python
for item in new:
  print(item)
```

```python
for item in new:
  print(item , new[item])
```

```python
for item, val in new.items():
  print(item, val)
```

```python
for key in new.keys():
  print(key)
```

```python
# Loop through only values
for xyz in new.values():
  print(xyz)
```

```python
# You can sort dictionary
for key in sorted(new):
  print(key)
```

Nested Disctionary

```python
company = {
    'name': 'Apple',
    'year': 1975,
    'founders': {
        'first': 'Steve Jobs',
        'second': 'Steve Wozniak'
    }
    }
print(company)
```

```python
print(company['founders'])
```

```python
print(company['founders']['first'])
```

```python
for item, val in company.items():
  print(item,val)
```
